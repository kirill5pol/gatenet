```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
sns.set_style("dark")
```

 
 ## Generate heatmap based on gate values


```python
test_data = np.random.rand(11,10) ** 10
test_data = (test_data.T / np.sum(test_data, axis=1)).T
#print(test_data)
#print(np.sum(test_data, axis=1))
```

```python
# Define the gate data here
```

```python
sns.heatmap(test_data, vmin=0, vmax=1)
```

```python
# To visualize how the gamma effects the gating
N = 20
seed_data = np.random.rand(11,10)

for i in range(1,N):
    test_data = seed_data ** i
    test_data = (test_data.T / np.sum(test_data, axis=1)).T
    plt.figure()
    sns.heatmap(test_data, vmin=0, vmax=1)
```
